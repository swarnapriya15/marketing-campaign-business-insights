**CONVERT DATE COLUMN**

```python
df["Date"]=pd.to_datetime(df["Date"],dayfirst=True)
```

```python
df.info()
```

**DESCRIPTIVE STATISTICS**

```python
df.describe()
```

```python
df["Channel_Used"] = df["Channel_Used"].str.split(",")

df = df.explode("Channel_Used")

df["Channel_Used"] = df["Channel_Used"].str.strip()
```

**VISUALIZE MARKETING CHANNELS**

```python
plt.figure(figsize=(8,5))

channel_count = df["Channel_Used"].value_counts()

plt.bar(channel_count.index, channel_count.values)

plt.title("Marketing Channels Distribution")
plt.xlabel("Marketing Channel")
plt.ylabel("Number of Campaigns")
plt.xticks(rotation=45)

plt.show()
```

 AVERAGE REVENUE BY CHANNEL

```python
plt.figure(figsize=(8,5))

df.groupby("Channel_Used")["Revenue"].mean().sort_values().plot(kind="bar")

plt.title("Average Revenue by Channel")
plt.xlabel("Marketing Channel")
plt.ylabel("Average Revenue")
plt.xticks(rotation=45)

plt.show()
```

**AVERAGE ROI BY CHANNEL**

```python
plt.figure(figsize=(8,5))

df.groupby("Channel_Used")["ROI"].mean().sort_values().plot(kind="bar")

plt.title("Average ROI by Channel")
plt.xlabel("Marketing Channel")
plt.ylabel("Average ROI")
plt.xticks(rotation=45)

plt.show()
```

**AVERAGE ACQUISITION COST BY CHANNEL**

```python
plt.figure(figsize=(8,5))

df.groupby("Channel_Used")["Acquisition_Cost"].mean().sort_values().plot(kind="bar")

plt.title("Average Acquisition Cost by Channel")
plt.xlabel("Marketing Channel")
plt.ylabel("Average Acquisition Cost")
plt.xticks(rotation=45)

plt.show()
```

**AVERAGE CONVERSIONS BY CHANNEL**

```python
plt.figure(figsize=(8,5))

df.groupby("Channel_Used")["Conversions"].mean().sort_values().plot(kind="bar")

plt.title("Average Conversions by Channel")
plt.xlabel("Marketing Channel")
plt.ylabel("Average Conversions")
plt.xticks(rotation=45)

plt.show()
```

**CORRELATION HEATMAP**

```python
plt.figure(figsize=(12,8))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()
```

**MARKETING FUNNEL ANALYSIS**

```python
funnel = df[["Impressions", "Clicks", "Leads", "Conversions"]].sum()

plt.figure(figsize=(8,5))

plt.bar(funnel.index, funnel.values)

plt.title("Marketing Funnel Analysis")
plt.xlabel("Funnel Stage")
plt.ylabel("Total Count")

plt.show()
```
